```python
img=cv.imread("/home/boss/Documents/Dataset_BUSI_with_GT/malignant/malignant (1).png",1)
# img=cv.imread("/home/boss/Downloads/frame extract/dataset/ankitframe0.jpg")
# smoothimg=wiener(img,mysize=5)
# nimg=np.array(smoothimg,dtype=np.uint8)
# imggray=cv.cvtColor(segment,cv.COLOR_BGR2GRAY)
# threshval,threshold=cv.threshold(imggray,0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)

imag=ID.noise_remMedianWeiner(img)
val=ID.CLAHEcontrastEnhance(imag)
img1=cv.cvtColor(val,cv.COLOR_RGB2GRAY)
thresval,threshimg=cv.threshold(img1,0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)

threshold1=cv.bitwise_not(threshimg)
kernel=np.ones((3,3),np.uint8)
morphimg=cv.morphologyEx(threshold1,cv.MORPH_OPEN,kernel,iterations=2)
bg2=cv.dilate(morphimg,kernel,iterations=1)
print(img.shape)
```

```python
dist_transform = cv.distanceTransform(morphimg,cv.DIST_L2,5)
ret, sure_fg = cv.threshold(dist_transform,0.7*dist_transform.max(),255,0)
sure_fg = np.uint8(sure_fg)
```

```python
unknown = np.subtract(bg2, sure_fg)
# cv.imshow('unknown', unknown)
# cv.imshow("thres",threshold1)
# cv.waitKey(0)
# cv.destroyAllWindows()
```

```python
ret, markers = cv.connectedComponents(sure_fg)
print(markers)
markers = markers +1
markers[unknown == 255] = 0
markers = cv.watershed(img, markers)
img[markers == -1] = (255, 0, 0)

cv.imshow("thres",threshold1)
cv.imshow('watershed',img)
cv.waitKey(0)
cv.destroyAllWindows()
```
