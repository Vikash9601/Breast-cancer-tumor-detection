```python
import numpy as np
import skfuzzy as fuzz
import cv2 as cv

def ifcm(data, num_clusters, m, error_threshold, max_iterations):
    # Initialize the membership and non-membership matrices randomly
    print(data.shape)
    U = np.random.rand(num_clusters,data.shape[-1])
    V = np.random.rand(num_clusters, data.shape[-1])

    # Normalize the membership and non-membership matrices
    U /= np.sum(U, axis=0)
    V /= np.sum(V, axis=0)
    
    # Iterate until convergence or reaching the maximum number of iterations
    for i in range(max_iterations):
        # Calculate the cluster centers
        centroids = fuzz.intuitionistic_fcm(data.reshape(-1, 1), U, V, num_clusters, m)

        # Calculate the membership and non-membership matrices
        U_new, V_new, _, _ = fuzz.intuitionistic_fcm(data.reshape(-1, 1), centroids, m=m)

        # Calculate the error
        error = np.linalg.norm(U_new - U) + np.linalg.norm(V_new - V)

        # Check for convergence
        if error < error_threshold:
            break

        U = U_new
        V = V_new

    # Return the final cluster centers and membership matrix
    return centroids, U

# Example usage
data = np.array([1, 2, 3, 4, 5, 6])  # Input data
num_clusters = 2  # Number of clusters
m = 2.0  # Fuzziness parameter
error_threshold = 0.001  # Error threshold for convergence
max_iterations = 100  # Maximum number of iterations



original = cv.imread("/home/boss/Documents/Dataset_BUSI_with_GT/malignant/malignant (1).png",0)

data=original.reshape(-1)
centroids, membership = ifcm(data, 3, 3.5, 0.12,30)
print("Cluster Centers:")
print(centroids)
print("Membership Matrix:")
print(membership)
```

```python

import cv2 as cv
import numpy as np
from skimage.restoration import (denoise_wavelet, estimate_sigma)
from skimage.metrics import peak_signal_noise_ratio
import import_ipynb
import ImageDenoise as ID
original = cv.imread("/home/boss/Documents/Dataset_BUSI_with_GT/malignant/malignant (1).png")

sigma = 0.12
sigma_est = estimate_sigma(original, channel_axis=-1, average_sigmas=True)

# original=np.array(org,dtype=np.float32)
im_bayes = denoise_wavelet(original, channel_axis=-1, convert2ycbcr=True,
                           method='BayesShrink', mode='soft',
                           rescale_sigma=True)
im_visushrink = denoise_wavelet(original, channel_axis=-1, convert2ycbcr=True,
                                method='VisuShrink', mode='soft',
                                sigma=sigma_est, rescale_sigma=True)

im_visushrink2 = denoise_wavelet(original, channel_axis=-1, convert2ycbcr=True,
                                 method='VisuShrink', mode='soft',
                                 sigma=sigma_est/2, rescale_sigma=True)
im_visushrink4 = denoise_wavelet(original, channel_axis=-1, convert2ycbcr=True,
                                 method='VisuShrink', mode='soft',
                                 sigma=sigma_est/4, rescale_sigma=True)

im_bayes=(im_bayes*255).astype(np.uint8)
# psnr_noisy = peak_signal_noise_ratio(original, noisy)
psnr_bayes = peak_signal_noise_ratio(original, im_bayes)
psnr_visushrink = peak_signal_noise_ratio(original, im_visushrink)
psnr_visushrink2 = peak_signal_noise_ratio(original, im_visushrink2)
psnr_visushrink4 = peak_signal_noise_ratio(original, im_visushrink4)

print(f"bayes:{psnr_bayes}, visushrink:{psnr_visushrink},visushrink1:{psnr_visushrink2},visushrink2:{psnr_visushrink4}")

image=ID.contrast_enhanceWMean(im_bayes)
image1=ID.contrast_enhanceWMedian(im_bayes)
cv.imshow("original",original)
cv.imshow("bayes",im_bayes)
cv.imshow("median",image)
cv.imshow("nlmean",image1)
cv.waitKey(0)
cv.destroyAllWindows()

```
